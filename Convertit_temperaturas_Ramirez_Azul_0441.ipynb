 Fórmulas de Conversión Para convertir entre estas dos escalas, utilizamos relaciones lineales basadas en los puntos de congelación (0°C / 32°F) y ebullición (100°C / 212°F) del agua.De Celsius a Fahrenheit:$$°F = (°C \times \frac{9}{5}) + 32$$(También se puede usar $1.8$ en lugar de $9/5$)De Fahrenheit a Celsius:$$°C = \frac{(°F - 32) \times 5}{9}$$(O bien, $(°F - 32) \div 1.8$)2. Ejemplos Resueltos Paso a Paso

Ejemplo A: Convertir 25°C a Fahrenheit

Identificar la fórmula: $°F = (°C \times 1.8) + 32$

Sustituir el valor: $°F = (25 \times 1.8) + 32$

Multiplicar: $25 \times 1.8 = 45$

Sumar 32: $45 + 32 = 77$

Resultado: 25°C equivalen a 77°F.

Ejemplo B: Convertir 98.6°F a Celsius

Identificar la fórmula: $°C = (°F - 32) / 1.8$

Sustituir el valor: $°C = (98.6 - 32) / 1.8$

Restar 32: $98.6 - 32 = 66.6$

Dividir entre 1.8: $66.6 / 1.8 = 37$

Resultado: 98.6°F equivalen a 37°C.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngresa los grados Celsius: "))
    f = (c * 9/5) + 32
    print(f"---> {c}°C equivalen a {f:.2f}°F")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngresa los grados Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"---> {f}°F equivalen a {c:.2f}°C")

def menu():
    """Muestra el menú y gestiona la lógica principal"""
    while True:
        print("\n--- CONVERSOR DE TEMPERATURA ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("Ramirez Azul NL = 44! ")
            break
        else:
            print("Opción no válida. Intenta de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
